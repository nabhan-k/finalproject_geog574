## Trains

```python
df = pd.read_csv("NYC_subway_traffic_2017-2021.csv")
df
```

```python
list(df.columns)
```

```python
#select main columns
main_df = df[["Unique ID", "Datetime", "Stop Name", "Line", "Connecting Lines", "Latitude", "Longitude",
              "Entries", "Exits"]]
main_df
```

```python
#editing columns for better analysis
main_df['Year'] = pd.DatetimeIndex(main_df['Datetime']).year
main_df["Num_Connecting_Lines"] = df['Connecting Lines'].str.len()
main_df
```

```python
#Only choosing 2017 and 2020
final = main_df.loc[(main_df['Year'] == 2017) | (main_df['Year'] == 2020)]
final
```

```python
#reselect main columns
final = final[["Unique ID", "Year", "Stop Name", "Line", "Connecting Lines","Num_Connecting_Lines" ,
                   "Latitude", "Longitude", "Entries", "Exits"]]
final
```

```python
final = final.rename(columns={'Unique ID': 'id', 'Year': 'year', 'Stop Name': 'subway_name',
                                  'Line': 'main_line', 'Connecting Lines': 'on_line',
                                  'Num_Connecting_Lines': 'num_on_line' })
final
```

```python
final.to_csv("trains_2017to2020.csv", sep=',', encoding='utf-8', index=False)
```
